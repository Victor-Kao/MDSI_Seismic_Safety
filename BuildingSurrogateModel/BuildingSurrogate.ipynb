```python
import numpy as np
import matplotlib.pyplot as plt

def compute_time_domain_response(acceleration_signal, fs, f_n, frf_amplitude):
    """
    Compute the time-domain acceleration response of an SDOF system using a single FRF point.
    
    Parameters:
    - acceleration_signal: Time-domain acceleration signal (1D array)
    - fs: Sampling frequency (Hz)
    - f_n: Natural frequency of the SDOF system (Hz)
    - frf_amplitude: Amplitude of the FRF at the natural frequency
    
    Returns:
    - reconstructed_signal: Reconstructed time-domain acceleration response
    """
    N = len(acceleration_signal)
    time = np.arange(N) / fs  # Time axis
    freqs = np.fft.rfftfreq(N, 1 / fs)  # Frequency axis
    fft_acc = np.fft.rfft(acceleration_signal)  # FFT of the signal
    
    # Find the index corresponding to the natural frequency
    idx_f_n = np.argmin(np.abs(freqs - f_n))
    
    # Amplitude and phase at the natural frequency
    #amplitude_at_f_n = np.abs(fft_acc[idx_f_n])
    amplitude_at_f_n = np.abs(fft_acc[idx_f_n])
    phase_at_f_n = np.angle(fft_acc[idx_f_n])
    
    # Scale the amplitude by the FRF
    scaled_amplitude = amplitude_at_f_n * frf_amplitude
    
    # Reconstruct the time-domain response for this frequency
    reconstructed_signal = scaled_amplitude * np.cos(2 * np.pi * f_n * time + phase_at_f_n)
    
    return reconstructed_signal

# Parameters
fs = 1000  # Sampling frequency (Hz)
T = 5      # Signal duration (s)
f_n = 5    # Natural frequency (Hz)
 # FRF amplitude at the natural frequency
damping_ratio = 0.04  # Damping ratio for SDOF FRF

# Generate a random force signal
#np.random.seed(42)
time = np.linspace(0, T, fs * T, endpoint=False)
force_signal = np.random.randn(len(time))  # Random force signal

# Compute the FFT of the force signal
N = len(force_signal)
freqs = np.fft.rfftfreq(N, 1 / fs)
fft_force = np.fft.rfft(force_signal)

# Whole SDOF FRF method
sdof_frf = 1 / np.sqrt((1 - (freqs / f_n) ** 2) ** 2 + (2 * damping_ratio * freqs / f_n) ** 2)
frf_amplitude = np.max(np.abs(sdof_frf))
print(frf_amplitude)
response_fft_whole = fft_force * sdof_frf
response_whole = np.fft.irfft(response_fft_whole)

# Single-point method
response_single_point = compute_time_domain_response(force_signal, fs, f_n, frf_amplitude/fs)

# Compare maximum accelerations
max_acc_whole = np.max(np.abs(response_whole))
max_acc_single_point = np.max(np.abs(response_single_point))

# Print results
print("Maximum acceleration (Whole FRF method):", max_acc_whole)
print("Maximum acceleration (Single-point method):", max_acc_single_point)

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(time, response_whole, label="Whole FRF Method", alpha=0.7)
plt.plot(time, response_single_point, label="Single-Point Method", linestyle="--")
plt.xlabel("Time (s)")
plt.ylabel("Acceleration")
plt.title("Time-Domain Responses Using Whole FRF and Single-Point Methods")
plt.legend()
plt.show()

```
